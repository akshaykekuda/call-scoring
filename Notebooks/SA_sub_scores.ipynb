```python
score_df = pd.DataFrame()#index = df['InteractionIdKey'].unique(), columns = q_text)
```

```python
for row in calls_df.itertuples():
    score_df.loc[row.InteractionIdKey,row.QuestionText] = row.AnswerScore
    score_df.loc[row.InteractionIdKey,'CombinedPercentileScore'] = row.CombinedPercentileScore
    if row.CombinedPercentileScore > 75:
        score_df.loc[row.InteractionIdKey,'Category'] = int(1)
    else:
        score_df.loc[row.InteractionIdKey,'Category'] = int(0)
```

```python
score_df[q_text] = score_df[q_text].astype('int64')
```

```python
import nltk
for row in df[df.Professionalism == 0].itertuples():
    txt = row.text
    allWords = nltk.tokenize.word_tokenize(txt)
    stopwords = nltk.corpus.stopwords.words('english')
    allWordExceptStopDist = nltk.FreqDist(w.lower() for w in allWords if ((w.lower() not in stopwords) and (w !='.')))
    print(allWordExceptStopDist.most_common(10))
```

```python
import re
import nltk
from collections import Counter
for row in df[df.Professionalism == 0].itertuples():
    txt = row.text
    allWords = nltk.tokenize.word_tokenize(txt)
    counts = Counter(allWords)
    if 'no' in allWords:
        print(re.findall(r"(.*\sno\s.*)",txt))
```
