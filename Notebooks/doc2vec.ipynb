```python
from gensim.models.doc2vec import Doc2Vec
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
import os
import gensim
train_file = "/mnt/transcriber/Call_Scoring/transcriptions/csr_ch/train/"
```

```python
import smart_open
def read_corpus(train_file, tokens_only=False):
    for file in os.listdir(train_file):
        with smart_open.open(train_file+file) as f:
            for i, line in enumerate(f):
                tokens = gensim.utils.simple_preprocess(line)
                if tokens_only:
                    yield tokens
                else:
                    # For training data, add tags
                    yield gensim.models.doc2vec.TaggedDocument(tokens, [i])

train_corpus = list(read_corpus(train_file))
```

```python

model = Doc2Vec(vector_size=256, window=2, min_count=1, workers=16)
```

```python
model.build_vocab(train_corpus)
```

```python
model.epochs
```

```python
model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
from gensim.test.utils import get_tmpfile
fname = get_tmpfile("my_doc2vec_model")
model.save(fname)
```

```python
model.save("/mnt/transcriber/word_embeddings/trained_doc2vec")
```
